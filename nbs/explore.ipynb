```python
place = pd.read_csv(DATA_DIR + "place", index_col=0)


def get_country_area_subarea(x):
    x = x.split('_')

    try:
        country = x[0]
    except:
        country = np.nan
    try: 
        area = x[1]
    except:
        area = np.nan 
    try:
        subarea = x[2]
    except:
        subarea = np.nan
    return [country, area, subarea]
    
area = [get_country_area_subarea(x) for x in place.place_name.values]


place_df = pd.DataFrame({'place_id':place.place_id,
                          'country' : [x[0] for x in area],
                          'area' :  [x[1] for x in area],
                          'subarea' : [x[2] for x in area]})
        
```
